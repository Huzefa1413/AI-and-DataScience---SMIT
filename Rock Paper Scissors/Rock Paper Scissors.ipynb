# Rock Paper Scissors

```python
import random
```

```python
user_points = 0
comp_points = 0
while True:
    print("---------------------------------------------------------------------------------")
    user_move = input("Enter a number between 1 and 3\n1: Rock\n2: Paper\n3: Scissor\n")
    comp_move = random.randint(1,3)
    if user_move == 1:
        if comp_move == 1:
            print("Draw")
        elif comp_move == 2:
            print("Computer Won")
            comp_points += 1
            if comp_points == 3:
                break
        else:
            print("You Won")
            user_points += 1
            if user_points == 3:
                break
    elif user_move == 2:
        if comp_move == 1:
            print("You Won")
            user_points += 1
            if user_points == 3:
                break
        elif comp_move == 2:
            print("Draw")
        else:
            print("Computer Won")
            comp_points += 1
            if comp_points == 3:
                break
    else:
        if comp_move == 1:
            print("Computer Won")
            comp_points += 1
            if comp_points == 3:
                break
        elif comp_move == 2:
            print("You Won")
            user_points += 1
            if user_points == 3:
                break
        else:
            print("Draw")
    print("Computer Points:",comp_points)
    print("User Points:",user_points)
if user_points > comp_points:
    print("You Won the Game")
else:
    print("Computer Won the Game")
```
