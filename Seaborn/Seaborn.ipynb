```python
# Import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")
tips
```

```python
# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
plt.show()
```

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
plt.show()
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
plt.show()
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
plt.show()
```

```python
sns.catplot(data=tips, kind="swarm", x="day", y="total_bill", hue="smoker")
plt.show()
```

```python
sns.catplot(data=tips, kind="violin", x="day", y="total_bill", hue="smoker", split=True)
plt.show()
```

```python
sns.catplot(data=tips, kind="bar", x="day", y="total_bill", hue="smoker")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from seaborn
titanic_df = sns.load_dataset("titanic")

# Filter the dataset to include only males
males_df = titanic_df[titanic_df["sex"] == "male"]

# Count the number of males who survived and died
survived = males_df[males_df["survived"] == 1].shape[0]
died = males_df[males_df["survived"] == 0].shape[0]

# Create a list of labels
labels = ["Survived", "Died"]

# Create a list of proportions
proportions = [survived, died]

# Create the pie chart
plt.pie(proportions, labels=labels, autopct="%1.1f%%")

# Add a title
plt.title("Proportion of Males Survived and Died")

# Display the plot
plt.show()
```
