# now open file with block

```python
with open("./fileone.txt") as f:
    f.seek(5) # change cursor position
    print(f.read())
```

```python
with open("./fileone.txt", mode="w") as f:
    f.write("HI, I am Huzefa")
```

```python
with open("./fileone.txt", mode="w+") as f:
    f.write("Hi, I am Huzefa")
    f.seek(0)
    f.read()
```

```python
with open("./fileone.txt", mode="r+") as f:
    f.write("Hello World")
    print(f.read())
```

```python
with open("./fileone.txt", mode="r+") as f:
    f.write("Hello World")
    f.seek(0)
    print(f.read())
```

```python
with open("./fileone.txt", mode="r+") as f:
    f.seek(len(f.read()))
    f.write("Hello World")
    f.seek(0)
    print(f.read())
```

```python
with open("./fileone2.txt", mode="w+") as f: # new file created
    f.seek(len(f.read()))
    f.write("Hello World")
    f.seek(0)
    print(f.read())
```

```python
with open("./fileone.txt", mode="rb") as f:
    print(type(f))
    print(f.read())
```

# Store input user in file using while loop

```python
while True:
    name = input("Enter Name:")
    if name.lower() == 'exit':
        break
    with open("./studentName.txt", "a") as f:
        f.write(name+"\n")
        
```
