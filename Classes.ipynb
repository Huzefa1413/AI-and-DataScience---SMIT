```python
class Employee:
    def _init_(self, id, name, basicSalary):
        self.id = id
        self.name = name
        self.basicSalary = basicSalary

    def get_id(self):
        return self.id

    def get_name(self):
        return self.name

    def get_basicSalary(self):
        return self.basicSalary 

    def set_id(self, id):
        self.id = id

    def set_name(self, name):
        self.id = name

    def set_basicSalary(self, basicSalary):
        self.id = basicSalary
    
    def display(self):
        return print(f"ID: {self.id} - Name: {self.name} - Salary: {self.basicSalary} PKR")

class Faculty(Employee):
    def _init_(self, id, name, basicSalary, degree, teachingHours):
        super().__init__(id,name,basicSalary)
        self.degree = degree
        self.teachingHours = teachingHours
        self.salary = basicSalary + teachingHours * 1000

    def get_degree(self):
        return self.degree

    def get_teachingHours(self):
        return self.teachingHours 

    def set_degree(self, degree):
        self.degree = degree

    def set_teachingHours(self, teachingHours):
        self.teachingHours = teachingHours
        self.salary = self.basicSalary + teachingHours * 1000

    def display(self):
        return print(f"ID: {self.id} - Name: {self.name} - Degree: {self.degree} - Teaching hours: {self.teachingHours} - Salary: {self.salary} PKR")

class Staff(Employee):
    def _init_(self, id, name, basicSalary, jobtitle, workingHours):
        super().__init__(id, name, basicSalary)
        self.jobtitle = jobtitle
        self.workingHours = workingHours
        self.salary = basicSalary
        if workingHours > 8:
            self.salary = basicSalary + (basicSalary * 25/100)

    def get_jobtitle(self):
        return self.jobtitle

    def get_workingHours(self):
        return self.workingHours 

    def set_jobtitle(self, jobtitle):
        self.jobtitle = jobtitle

    def set_workingHours(self, workingHours):
        self.workingHours = workingHours
        self.salary = self.basicSalary
        if workingHours > 8:
            self.salary = self.basicSalary + (self.basicSalary * 25/100) 

    def display(self):
        return print(f"ID: {self.id} - Name: {self.name} - Job Title: {self.jobtitle} - Working hours: {self.workingHours} - Salary: {self.salary} PKR")

Huzefa = Employee(96, "Huzefa", 100000)
Bilal = Faculty(52, "Bilal", 100000, "BSCS", 16)
Saim = Staff(91, "Saim", 100000, "Admin", 9)
Huzefa.display()
Bilal.display()
Saim.display()
```
